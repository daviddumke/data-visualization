# One of the simplest plots you can make in Matplotlib

```python
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 22, 25]

# plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100)
ax.plot(x_values, y_values, linewidth=3)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x_values = range(1, 26)
y_values = [x**random.randint(2, 3) for x in x_values]
random.shuffle(y_values)

# plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=50, c='r', alpha=0.5)

ax.plot(x_values, y_values, linewidth=3)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', which='major', labelsize=14)
# ax.axis([0, 25, 0, 100000])
plt.show()
```
