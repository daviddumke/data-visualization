```python
import matplotlib.pyplot as plt

from random_walk import RandomWalk


# Make a random walk.
rw = RandomWalk(50_000)
rw.fill_walk()

# Plot the walk.
fig, ax = plt.subplots(figsize=(15, 9))
point_numbers = range(rw.num_points)

ax.scatter(rw.x_values, rw.y_values, s=5, c=point_numbers, alpha=0.5, 
        cmap=plt.cm.Blues, edgecolors='none')

# Remove the axes.
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

# Emphasize the first and last points.
ax.scatter(0, 0 , c='green', edgecolors='none', s=10)
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=10)

plt.show()

```
